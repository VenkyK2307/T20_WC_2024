# BATTING

## Scraping The DATA

```python
import pandas as pd
```

## Scrapped Data from ESPN and CricBuzz Websites

```python
data = pd.read_html("https://www.cricbuzz.com/cricket-series/7476/icc-mens-t20-world-cup-2024/stats")[0]
data
```

```python

data1 = pd.read_csv("batting_stats.csv")
data1
```

## Data Cleaning

### Replacing "-" to 0's

```python
data1.replace("-", 0, inplace=True)
data1.head(10)
```

### Replacing "HS" Scores (example 83* --> 83)

```python

data1['HS'] = data1['HS'].astype(str).str.replace('*', '', regex=False)

data1['HS'] = pd.to_numeric(data1['HS'], errors='coerce')

data1.head(10)
```

## Saving as CSV File


```python
data1.to_csv("batting_stats.csv", index=False)
```

## Filtering Best Batsman Based on Conditions

### Conditions 
1) Batting average >  35 Runs
2) Strike Rate > 140
3) High score > 70
4) Innings Played > 3
5) Boundary Ratio > 40%


```python

# Now apply the filter conditions
condition = (
    (data1['Ave'] > 35) &
    (data1['SR'] > 140) &
    (data1['HS'] > 70) &
    (data1['Inns'] > 3) &
    (((data1['4s'] * 4 + data1['6s'] * 6) / data1['Runs']) * 100 > 40) 
    
)
filtered_players_bat = data1[condition]
filtered_players_bat
```

# BOWLING

## Scraping data

```python

data2 = pd.read_csv("bowling_stats.csv")
data2
```

## Cleaning the data 

### Replacing - to 0's

```python
data2.replace("-", 0, inplace=True)
data2.head(10)
```

## Saving as CSV


```python
data2.to_csv("bowling_stats.csv", index=False)
```

## Filtering Bowlers Based on Conditions

### Conditions 
1) Innings palyed  >  4 
2) Wickets Taken > 11
3) Economy < 7 (Average runs given for 1 Over)
4) Strike Rate < 16 (No. of balls to take a wicket)
5) Average < 20 (Runs given per wicket)


```python
# Apply filter conditions
filtered_players_bowl = data2[
    (data2['Inns'] > 4) & 
    (data2['Econ'] < 7) & 
    (data2['Wkts'] > 11) & 
    (data2['SR'] < 13) & 
    (data2['Ave'] < 20)
]

filtered_players_bowl
```

# TOP 11 PLAYERS OF THE TOURNAMENT

```python
filtered_players_bat = filtered_players_bat[['Player', 'Span']]
filtered_players_bowl = filtered_players_bowl[['Player', 'Span']]

filtered_players_combined = pd.concat([filtered_players_bat, filtered_players_bowl], ignore_index=True)

filtered_players_combined
```
